```python
from rl_qlearning import Q_learning
import matplotlib.pyplot as plt
import gym
%matplotlib inline

# initialize the environment
env = gym.make('CartPole-v1')
```

```python
n = 50 # number of episodes to average

# simulate the Q-learning
rl_qlearning1 = Q_learning(env)
rl_qlearning1.episode = 500
rl_qlearning1.gamma = 0.9 # discount factor
rl_qlearning1.alpha = 0.1 # learning rate
rl_qlearning1.epsilon = 1 # exploration rate
rl_qlearning1.buckets = 2 # number of buckets
rl_qlearning1.simulate()
y1 = rl_qlearning1.reward_history

rl_qlearning2 = Q_learning(env)
rl_qlearning2.episode = 500
rl_qlearning2.gamma = 0.9 # discount factor
rl_qlearning2.alpha = 0.1 # learning rate
rl_qlearning2.epsilon = 1 # exploration rate
rl_qlearning2.buckets = 5 # number of buckets
rl_qlearning2.simulate()
y2 = rl_qlearning2.reward_history

rl_qlearning3 = Q_learning(env)
rl_qlearning3.episode = 500
rl_qlearning3.gamma = 0.9 # discount factor
rl_qlearning3.alpha = 0.1 # learning rate
rl_qlearning3.epsilon = 1 # exploration rate
rl_qlearning3.buckets = 10 # number of buckets
rl_qlearning3.simulate()
y3 = rl_qlearning3.reward_history

y1 = [sum(y1[i:i+n])/n for i in range(0, len(y1), n)]
x1 = range(0,len(y1)*n,n)

y2 = [sum(y2[i:i+n])/n for i in range(0, len(y2), n)]
x2 = range(0,len(y2)*n,n)

y3 = [sum(y3[i:i+n])/n for i in range(0, len(y3), n)]
x3 = range(0,len(y3)*n,n)

# plot the graph
plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3)
plt.legend(['buckets = 2', 'buckets = 5', 'buckets = 10'])
plt.xlabel('Episode')
plt.ylabel('Sum of Rewards in Episode')
plt.title('a) State Space size')
# save the figure
plt.savefig('graph/graph a.png')
```
